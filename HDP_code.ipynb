```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
import numpy as np

# User input with correct types
age = int(input("Age: "))
cp = int(input("Chest Pain Type (cp) [0-3]: "))
trestbps = float(input("Resting Blood Pressure (trestbps): "))
chol = float(input("Serum Cholesterol (chol): "))
fbs = int(input("Fasting Blood Sugar > 120 mg/dl? (fbs) [0 = No, 1 = Yes]: "))
restecg = int(input("Resting ECG Results (restecg) [0-2]: "))
thalach = float(input("Maximum Heart Rate Achieved (thalach): "))
exang = int(input("Exercise Induced Angina (exang) [0 = No, 1 = Yes]: "))
oldpeak = float(input("ST Depression Induced (oldpeak): "))
slope = int(input("Slope of the ST Segment (slope) [0-2]: "))
ca = int(input("Number of Major Vessels (ca) [0-3]: "))
thal = int(input("Thalassemia (thal) [1 = Normal, 2 = Fixed Defect, 3 = Reversible Defect]: "))
sex_input = input("Sex (male/female): ").strip().lower()
sex_male = 1 if sex_input == "male" else 0

# Combine into a NumPy array for prediction (exclude target if predicting)
user_input = np.array([[age, cp, trestbps, chol, fbs, restecg, thalach,
                        exang, oldpeak, slope, ca, thal, sex_male]])



```

```python
user_input
```

```python
# Convert to DataFrame
# Column names must match the features used in training
columns = ['age', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
           'exang', 'oldpeak', 'slope', 'ca', 'thal', 'sex_male']

# Correct: pass user_input directly, no extra brackets
input_df = pd.DataFrame(user_input, columns=columns)
# Apply the same scaler used for training
input_scaled = sc_input.transform(input_df)  # sc_input is the scaler used for X_train
```

```python
# Predict
prediction = classifier.predict(input_scaled)

# Output result
print("Prediction:", "Heart Disease Have" if prediction[0] == 1 else "No Heart Disease")
```
